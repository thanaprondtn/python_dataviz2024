```python
# prompt: ค้นหาความคิดเห็นที่มีจำนวนยอดไลค์ หรือจำนวนการตอบกลับ มากที่สุด แสดงผลเป็นตาราง

# ค้นหาความคิดเห็นที่มีจำนวนยอดไลค์ หรือจำนวนการตอบกลับ มากที่สุด
max_likes = test_df['ยอดไลค์'].max()
max_replies = test_df['จำนวนการตอบกลับ'].max()

most_liked_comments = test_df[test_df['ยอดไลค์'] == max_likes]
most_replied_comments = test_df[test_df['จำนวนการตอบกลับ'] == max_replies]

# รวมตารางของความคิดเห็นที่มีจำนวนยอดไลค์ และจำนวนการตอบกลับ มากที่สุด
result_df = pd.concat([most_liked_comments, most_replied_comments])

# แสดงผลเป็นตาราง
print("ความคิดเห็นที่มีจำนวนยอดไลค์ หรือจำนวนการตอบกลับ มากที่สุด:")
result_df
```

```python
import pandas as pd
from collections import Counter

# นับจำนวนครั้งที่แต่ละชนิดของมะเร็งถูกกล่าวถึง
all_words = [word for comment in test_df['comments'].str.lower().str.split() for word in comment]
cancer_counts = Counter(all_words)

# คัดกรองเฉพาะคำที่เกี่ยวข้องกับโรคมะเร็ง
cancer_keywords = ["มะเร็ง", "cancer"]
cancer_names = [word for word in cancer_counts if any(keyword in word for keyword in cancer_keywords)]

# หาชนิดของมะเร็งที่ถูกกล่าวถึงมากที่สุด
most_common_cancer = max(cancer_names, key=cancer_counts.get)
count = cancer_counts[most_common_cancer]

# สร้าง DataFrame สำหรับแสดงผล
result_df = pd.DataFrame({
    'มะเร็งชนิดที่ถูกกล่าวถึงมากที่สุด': [most_common_cancer],
    'จำนวนครั้ง': [count]
})

# แสดงผลเป็นตาราง
print("มะเร็งชนิดที่ถูกกล่าวถึงมากที่สุด:")
result_df
```
