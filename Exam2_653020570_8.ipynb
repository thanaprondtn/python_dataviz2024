```python
import os

# Replace with your actual folder path in Google Drive
path = '/content/drive/MyDrive/Data_Viz_2024_Data570-8'

# List of CSV file names
file_names = ['crime_scene_report.csv', 'drivers_license.csv', 'interview.csv', 'income.csv', 'get_fit_now_member.csv', 'get_fit_now_check_in.csv', 'facebook_event_checkin.csv', 'person.csv']

# Create an empty list to store dataframes
dataframes = []

# Loop through each file and read it into a dataframe
for file_name in file_names:
  file_path = os.path.join(path, file_name)
  df = pd.read_csv(file_path)
  dataframes.append(df)

# Now you have a list of dataframes that you can work with
# For example, print the first dataframe
print(dataframes[0])
```

```python
# prompt: จาก crime_scene_report.csv ให้เลือก type ที่เป็น murder และ date คือ 20180115 และ city คือ SQL City

# Assuming 'crime_scene_report.csv' is the first file in the list
crime_df = dataframes[0]

# Filter the dataframe
filtered_df = crime_df[(crime_df['type'] == 'murder') & (crime_df['date'] == 20180115) & (crime_df['city'] == 'SQL City')]
filtered_df
```

```python
# prompt: who lives at the last house on "Northwestern Dr"

# Assuming 'person.csv' is the last file in the list
person_df = dataframes[-1]

# Filter the dataframe for residents on Northwestern Dr, sort by address number, and get the last row
last_resident = person_df[person_df['address_street_name'] == 'Northwestern Dr'].sort_values('address_number', ascending=True).iloc[-1]

last_resident
```

```python
# prompt: หา คนที่ชื่อมีคำว่า Annabel และ ที่อยู่มีคำว่า Franklin

# Assuming 'person.csv' is the last file in the list
person_df = dataframes[-1]

# Filter the dataframe for names containing "Annabel" and address containing "Franklin"
filtered_person_df = person_df[(person_df['name'].str.contains("Annabel")) & (person_df['address_street_name'].str.contains("Franklin"))]
filtered_person_df
```

```python
# prompt: หา interview ของ id 14887 และ 16371

# Assuming 'interview.csv' is the third file in the list
interview_df = dataframes[2]

# Filter the dataframe for interview ids 14887 and 16371
interview_filtered = interview_df[interview_df['person_id'].isin([14887, 16371])]
interview_filtered
```

```python
# เชื่อมตาราง person กับ  drivers_license
person_df = dataframes[-1]
drivers_license_df = dataframes[1]

# Merge the two dataframes on the 'id' column in person and 'person_id' column in drivers_license
person_data = pd.merge(person_df, drivers_license_df, left_on='license_id', right_on='id', how='inner')

person_data
```

```python
# เชื่อมตาราง person_data กับ  get_fit_now_member
get_fit_now_member_df = dataframes[4]

test_df = pd.merge(person_data, get_fit_now_member_df, left_on='id_x', right_on='person_id', how='inner')

test_df
```

```python
result_df = test_df[(test_df['id'].str.contains('48Z')) & (test_df['membership_status'] == 'gold') & (test_df['plate_number'].str.contains('H42W'))]

result_df
```

```python
# เชื่อมตาราง test_df กับ  get_fit_now_check_in
get_fit_now_check_in_df = dataframes[5]

test2_df = pd.merge(test_df, get_fit_now_check_in_df, left_on='id', right_on='membership_id')

test2_df
```
