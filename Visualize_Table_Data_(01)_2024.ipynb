```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates
เอาคอลัมภ์ที่สนใจมาเรียงต่อกัน ขึ้นลง


```
pd.plotting.parallel_coordinates(df, 'Name');
```

## ปรับ parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#FFD700', '#C0C0C0','#CD7F32')) #html color
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r', 'g', 'b')) #ตัวย่อของสี R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.8,0), (0,0.7,0), (0.6,0,1))) #rgb value (r,g,b) | r,g,b -> [0,1] 1คือเปิดสว่างสุด 0 คือปิด
```

## ปรับ parameter: colormap

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='Pastel2')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

# เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy() #คอลัมภ์สุดท้ายเป็นชื่อเลยไม่เอา
np_data
```

# Visualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของ figure (ตัวเลขมากสีสว่าง ตัวเลขน้อยสีเข้ม)

plt.imshow(np_data[:50,:],interpolation='nearest')
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest')
```

# การใช้ subplot

```
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(np_data[:50,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,2)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma')
```

```
plt.imshow(np_data[:10,:],interpolation='nearest',cmap='magma',alpha = 0.5 ) #alpha chanel โปร่งใสแค่ไหน น้ยอคือโปร่งใส
```
