#เตรียมข้อมูล

สร้าง Column เพิ่มเติม

1. แยก conutry ออกเป็น country1, country2, ..., conutry และสร้างคอลัมน์ continent1, continent2, ..., continent n
2. Rate_Group --> แบ่งกลุ่ม Rate ออกเป็น 4 กลุ่ม
  - General_Audience (กลุ่มสำหรับผู้ชมทั่วไป)
    - G
    - TV-G
    - TV-Y
    - TV-Y7
  - Parental_Guidance (กลุ่มสำหรับผู้ปกครองแนะนำ)
    - PG
    - PG-13
    - TV-PG
    - TV-Y7-FV
    - TV-14
  - Mature Audiences (กลุ่มสำหรับผู้ใหญ่)
    - R
    - NC-17
    - TV-MA
  - Unrated/Not Rated (เนื้อหาที่ไม่ได้รับการจัดเรท)
    - UR
    - NR
3. Era --> ดูจาก release_year แบ่งออกเป็นยุค
4. duration_min --> แยกจาก duration สำหรับ Movie ส่วน TV Show ให้เป็น No available
5. duration_season --> แยกจาก duration สำหรับ TV Show ส่วน Movie ให้เป็น No available
6. year_added --> แยกปีจาก date_added
7. แยก Genre ออกเป็น Genre1, Genre2, ..., Gnere n

```python
!pip install pycountry_convert
import pandas as pd
import pycountry_convert as pc
from datetime import datetime

def get_rate_group(rate):
    general_audience = ['G', 'TV-G', 'TV-Y', 'TV-Y7']
    parental_guidance = ['PG', 'PG-13', 'TV-PG', 'TV-Y7-FV', 'TV-14']
    mature_audiences = ['R', 'NC-17', 'TV-MA']
    Unrated = ['UR', 'NR']

    if rate in general_audience:
        return 'General_Audience'
    elif rate in parental_guidance:
        return 'Parental_Guidance'
    elif rate in mature_audiences:
        return 'Mature_Audiences'
    elif rate in Unrated:
        return 'Unrated'
    else:
        return 'Unknown'

def get_era(year):
    if 1980 <= year <= 1989:
        return '80s'
    elif 1990 <= year <= 1999:
        return '90s'
    elif 2000 <= year <= 2009:
        return '2000s'
    elif 2010 <= year <= 2019:
        return '2010s'
    elif 2020 <= year <= 2029:
        return '2020s'
    elif year < 1980:
        return 'Pre-80s'
    else:
        return 'Post-2029'

def parse_duration(duration, content_type):
    if pd.isna(duration):
        return 'No available'
    if content_type == 'Movie':
        return int(duration.split()[0]) if 'min' in duration else 'No available'
    elif content_type == 'TV Show':
        return int(duration.split()[0]) if 'Season' in duration or 'Seasons' in duration else 'No available'
    else:
        return 'No available'

def extract_year(date_str):
    if pd.isna(date_str):
        return None
    date_str = str(date_str).strip()  # แปลงเป็น string และตัดช่องว่าง
    formats = ['%B %d, %Y', '%Y-%m-%d', '%d-%m-%Y', '%Y']  # เพิ่มรูปแบบวันที่ที่เป็นไปได้
    for fmt in formats:
        try:
            return datetime.strptime(date_str, fmt).year
        except ValueError:
            continue
    print(f"ไม่สามารถแปลงวันที่ได้: {date_str}")  # แสดงค่าที่ไม่สามารถแปลงได้
    return None

# Rate_Group
netflix_df['Rate_Group'] = netflix_df['rate'].apply(get_rate_group)

# Era
netflix_df['Era'] = netflix_df['release_year'].apply(get_era)

# Duration_min and Duration_season
netflix_df['duration_min'] = netflix_df.apply(lambda x: parse_duration(x['duration'], x['type']) if x['type'] == 'Movie' else 'No available', axis=1)
netflix_df['duration_season'] = netflix_df.apply(lambda x: parse_duration(x['duration'], x['type']) if x['type'] == 'TV Show' else 'No available', axis=1)

# Year_added
netflix_df['Year_added'] = netflix_df['date_added'].apply(extract_year)
```

## กราฟที่ 1 Radar Chart เปรียบเทียบผู้กำกับ Movie และผู้กำกับ TV Show
โดยมีตัวชี้วัด ดังนี้
1. Content_Count
2. Genre_Count
3. Rate_Count
4. Country_Count
5. Top_Actor_Count

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def count_top_actors(content_type):
    content_df = netflix_df[netflix_df['type'] == content_type]
    # รวบรวมรายชื่อนักแสดงทั้งหมด
    all_actors = [actor.strip() for actors in content_df['actor'].dropna() for actor in actors.split(',')]
    # นับความถี่ของแต่ละนักแสดง
    actor_counts = pd.Series(all_actors).value_counts()
    # กำหนด threshold ที่ percentile 75
    threshold = actor_counts.quantile(0.75)
    # นับจำนวน Top Actor
    top_actors = actor_counts[actor_counts > threshold]
    return len(top_actors), top_actors

def calculate_content_summary(content_type):
    content_df = netflix_df[netflix_df['type'] == content_type]
    content_count = content_df['title'].count()
    genre_count = len(set(','.join(content_df['genres'].fillna('')).split(',')))
    rate_count = content_df['rate'].nunique()
    country_count = len(set(','.join(content_df['country'].fillna('')).split(',')))
    top_actor_count, top_actors = count_top_actors(content_type)

    return pd.Series({
        'Content_Count': content_count,
        'Genre_Count': genre_count,
        'Rate_Count': rate_count,
        'Country_Count': country_count,
        'Top_Actor_Count': top_actor_count
    }), top_actors

# คำนวณสรุปข้อมูลสำหรับ Movie และ TV Show
movie_summary, movie_top_actors = calculate_content_summary('Movie')
tv_show_summary, tv_show_top_actors = calculate_content_summary('TV Show')

# สร้างตารางสรุป
summary_df = pd.DataFrame({'Movie': movie_summary, 'TV Show': tv_show_summary}).T

print("Summary Table:")
summary_df
```

```python
# สร้าง Radar Chart จากข้อมูลดิบ
categories = ['Content_Count', 'Genre_Count', 'Rate_Count', 'Country_Count', 'Top_Actor_Count']
N = len(categories)

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(12, 8), subplot_kw=dict(polar=True))

colors = ['#FF9999', '#66B2FF']

for i, content_type in enumerate(summary_df.index):
    values = summary_df.loc[content_type, categories].values.flatten().tolist()
    values += values[:1]

    ax.plot(angles, values, 'o-', linewidth=2, color=colors[i], label=content_type)
    ax.fill(angles, values, alpha=0.25, color=colors[i])

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, size=10)

ax.set_ylim(0, max(summary_df[categories].max()) * 1.1)

yticks = ax.get_yticks()
ax.set_yticks(yticks)
for y in yticks[1:-1]:
    plt.polar(angles, [y] * len(angles), '--', color='grey', alpha=0.3)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title("Director of Movies vs Director of TV Shows", size=15, y=1.08)
plt.tight_layout()
plt.show()
```

## กราฟที่ 3 Sunburst Chart

```python
import plotly.express as px
```

```python
# สร้างฟังก์ชันเพื่อรวม continent ทั้งหมด
def combine_continents(row):
    continents = [row[f'continent{i}'] for i in range(1, 13) if pd.notna(row[f'continent{i}'])]
    return ', '.join(set(continents))  # ใช้ set เพื่อลบค่าซ้ำ

# สร้างคอลัมน์ใหม่ที่รวม continent ทั้งหมด
netflix_df_merge['All_Continents'] = netflix_df_merge.apply(combine_continents, axis=1)

# สร้างกลุ่มข้อมูลสำหรับ Sunburst Chart
sunburst_data = netflix_df_merge.groupby(['Year_added', 'type', 'Rate_Group', 'All_Continents']).size().reset_index(name='count')

# แยกคอลัมน์ All_Continents เป็นหลายแถว
sunburst_data = sunburst_data.assign(Continent=sunburst_data['All_Continents'].str.split(', ')).explode('Continent')

# สร้าง Sunburst Chart
fig = px.sunburst(
    sunburst_data,
    path=['Year_added', 'type', 'Rate_Group', 'Continent'],
    values='count',
    title='Netflix Content Distribution by Year, Type, Rate_Group, and Continent',
    color='Year_added',
    color_continuous_scale='Viridis'
)

# ปรับแต่งการแสดงผล
fig.update_layout(
    width=1000,
    height=1000,
    title_font_size=24,
    title_x=0.5,
)

# แสดงกราฟ
fig.show()
```

## กราฟที่ 4 Stacked Area Chart

```python
import plotly.express as px
import pandas as pd
```

```python
# สร้างกลุ่มข้อมูลสำหรับ Stacked Area Chart
stacked_data = netflix_df.groupby(['Era', 'type']).size().reset_index(name='count')

# เรียงลำดับ Era ให้ถูกต้อง
era_order = ['Pre-80s', '80s', '90s', '2000s', '2010s', '2020s', 'Post-2029']
stacked_data['Era'] = pd.Categorical(stacked_data['Era'], categories=era_order, ordered=True)
stacked_data = stacked_data.sort_values('Era')

# สร้าง Stacked Area Chart
fig = px.area(stacked_data, x='Era', y='count', color='type',
              title='Netflix Content Distribution by Era and Type',
              labels={'count': 'Number of Titles', 'type': 'Content Type'},
              color_discrete_map={'Movie': '#FF9999', 'TV Show': '#66B2FF'})

# ปรับแต่งการแสดงผล
fig.update_layout(
    xaxis_title='Era',
    yaxis_title='Number of Titles',
    legend_title='Content Type',
    title_x=0.5,
    width=1000,
    height=600,
)

# แสดงกราฟ
fig.show()
```

## กราฟที่ 5 Bubble Chart

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# กรองข้อมูลสำหรับ Movie และ TV Show
movies_df = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['duration_min'] != 'No available')].copy()
movies_df['duration_min'] = pd.to_numeric(movies_df['duration_min'], errors='coerce')

tv_shows_df = netflix_df[(netflix_df['type'] == 'TV Show') & (netflix_df['duration_season'] != 'No available')].copy()
tv_shows_df['duration_season'] = pd.to_numeric(tv_shows_df['duration_season'], errors='coerce')

# แปลง Rate_Group เป็นค่าตัวเลข
rate_group_map = {'General_Audience': 0, 'Parental_Guidance': 1, 'Mature_Audiences': 2, 'Unrated': 3}
movies_df['Rate_Group_Num'] = movies_df['Rate_Group'].map(rate_group_map)
tv_shows_df['Rate_Group_Num'] = tv_shows_df['Rate_Group'].map(rate_group_map)

# กำหนดลำดับของ Era ที่ต้องการ
era_order = ['Pre-80s', '80s', '90s', '2000s', '2010s', '2020s']

# แปลงคอลัมน์ Era เป็นลำดับตามที่กำหนดสำหรับ movies_df
movies_df['Era'] = pd.Categorical(movies_df['Era'], categories=era_order, ordered=True)

# แปลงคอลัมน์ Era เป็นลำดับตามที่กำหนดสำหรับ tv_shows_df
tv_shows_df['Era'] = pd.Categorical(tv_shows_df['Era'], categories=era_order, ordered=True)

# สร้าง subplot
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=("Movie Duration by Era", "TV Show Seasons by Era"),
                    horizontal_spacing=0.2)

# สร้าง Bubble Chart สำหรับ Movies
fig.add_trace(
    go.Scatter(
        x=movies_df['Era'],
        y=movies_df['duration_min'],
        mode='markers',
        marker=dict(
            size=movies_df['duration_min'],
            sizemode='area',
            sizeref=2.*max(movies_df['duration_min'])/(40.**2),
            sizemin=3,
            color=movies_df['Rate_Group_Num'],
            colorscale='Viridis',
            showscale=True,
            colorbar=dict(
                title="Rate Group",
                tickvals=[0, 1, 2, 3],
                ticktext=['General', 'Parental', 'Mature', 'Unrated'],
                x=0.43,
                len=0.75
            )
        ),
        text=movies_df['title'],
        hovertemplate='<b>%{text}</b><br>Era: %{x}<br>Duration: %{y} min<br>Rate Group: %{marker.color}<extra></extra>'
    ),
    row=1, col=1
)

# สร้าง Bubble Chart สำหรับ TV Shows
fig.add_trace(
    go.Scatter(
        x=tv_shows_df['Era'],
        y=tv_shows_df['duration_season'],
        mode='markers',
        marker=dict(
            size=tv_shows_df['duration_season'],
            sizemode='area',
            sizeref=2.*max(tv_shows_df['duration_season'])/(40.**2),
            sizemin=3,
            color=tv_shows_df['Rate_Group_Num'],
            colorscale='Viridis',
            showscale=True,
            colorbar=dict(
                title="Rate Group",
                tickvals=[0, 1, 2, 3],
                ticktext=['General', 'Parental', 'Mature', 'Unrated'],
                x=1.02,
                len=0.75
            )
        ),
        text=tv_shows_df['title'],
        hovertemplate='<b>%{text}</b><br>Era: %{x}<br>Seasons: %{y}<br>Rate Group: %{marker.color}<extra></extra>'
    ),
    row=1, col=2
)

# ปรับแต่งการแสดงผล
fig.update_layout(
    title='Netflix Content: Movie Duration and TV Show Seasons by Era',
    height=700,
    width=1600,
    showlegend=False
)

fig.update_xaxes(title_text="Era", row=1, col=1, categoryorder='array', categoryarray=era_order)
fig.update_xaxes(title_text="Era", row=1, col=2, categoryorder='array', categoryarray=era_order)
fig.update_yaxes(title_text="Duration (minutes)", row=1, col=1)
fig.update_yaxes(title_text="Number of Seasons", row=1, col=2)

# ปรับขอบเขตของแกน y แยกสำหรับแต่ละกราฟ
fig.update_yaxes(range=[0, movies_df['duration_min'].max() * 1.1], row=1, col=1)
fig.update_yaxes(range=[0, 20], row=1, col=2)  # กำหนดค่าสูงสุดของแกน Y สำหรับ TV Show เป็น 20

# แสดงกราฟ
fig.show()
```
