* (10 points) จากข้อมูลที่ให้มา มีค่าว่างสูญหายมากน้อยแค่ไหน? และมีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
# ตรวจสอบค่าว่าง
missing_values = titanic.isnull().sum()
print("Missing Values:\n", missing_values)

# ตรวจสอบค่าผิดปกติ (outliers) ด้วย boxplot
plt.figure(figsize=(10, 6))
titanic.boxplot()
plt.xticks(rotation=45)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
# แทนที่ค่าว่างใน age ด้วยค่ามัธยฐาน เนื่องจากเป็นข้อมูลเชิงปริมาณและการกระจายของข้อมูลไม่เบ้มาก
titanic['age'].fillna(titanic['age'].median(), inplace=True)
```

```python
# ลบคอลัมน์ cabin เนื่องจากมีค่าว่างจำนวนมาก และยากที่จะแทนที่ด้วยค่าที่เหมาะสม
titanic.drop('cabin', axis=1, inplace=True)
```

```python
# แทนที่ค่าว่างใน embarked ด้วยค่าที่พบบ่อยที่สุด  เนื่องจากเป็นข้อมูลเชิงคุณภาพ
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)
```

```python
#แทนที่ค่าว่างด้วยค่ามัธยฐาน (median) เนื่องจากค่าโดยสารอาจมีการกระจายที่เบ้ (skewed) และค่ามัธยฐานจะไม่ถูกกระทบจากค่าผิดปกติมากเท่าค่าเฉลี่ย
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)
```

```python
#ค่า Boat เป็นข้อมูลเชิงคุณภาพที่แสดงถึงหมายเลขเรือชูชีพที่ผู้โดยสารใช้ ค่าว่างในคอลัมน์นี้อาจหมายความว่าผู้โดยสารไม่ได้ใช้เรือชูชีพ
#แทนที่ค่าว่างด้วยสตริง "Unknown" เพื่อระบุว่าข้อมูลไม่ทราบ
titanic['boat'].fillna("Unknown", inplace=True)
```

```python
#ค่า Body เป็นข้อมูลเชิงปริมาณที่แสดงถึงหมายเลขร่างของผู้โดยสารที่เสียชีวิต ค่าว่างในคอลัมน์นี้อาจหมายความว่าไม่พบร่างของผู้โดยสาร
#เนื่องจากคอลัมน์นี้มีค่าว่างจำนวนมาก การลบคอลัมน์นี้จึงทางเลือกที่ดีกว่าการพยายามแทนที่ค่าว่าง
titanic.drop('body', axis=1, inplace=True)
```

```python
#ค่า Home.dest เป็นข้อมูลเชิงคุณภาพที่แสดงถึงบ้านหรือจุดหมายปลายทางของผู้โดยสาร
#แทนที่ค่าว่างด้วยสตริง "Unknown" เพื่อระบุว่าข้อมูลไม่ทราบ
titanic['home.dest'].fillna("Unknown", inplace=True)
```

```python
titanic
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# ทำ one-hot encoding
embarked_encoded = pd.get_dummies(titanic['embarked'], prefix='embarked')

# แปลง True เป็น 1 และ False เป็น 0
embarked_encoded = embarked_encoded.astype(int)

# รวมข้อมูลที่เข้ารหัสแล้วเข้ากับ DataFrame เดิม
titanic = pd.concat([titanic, embarked_encoded], axis=1)

# ลบคอลัมน์ Embarked เดิม
titanic.drop('embarked', axis=1, inplace=True)
```

```python
# ใช้ .map() เพื่อแปลงเพศชายเป็น 0 และเพศหญิงเป็น 1
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
```

```python
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']
```

```python
titanic['IsAlone'] = np.where(titanic['FamilySize'] == 0, 1, 0)
```

```python
titanic
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# สร้าง MinMaxScaler object
scaler = MinMaxScaler()

# เลือกคอลัมน์ Fare และ Age
fare_age = titanic[['fare', 'age']]

# ทำ Min-Max Scaling
scaled_fare_age = scaler.fit_transform(fare_age)

# สร้าง DataFrame ใหม่จากข้อมูลที่ถูกปรับขนาด
scaled_fare_age_df = pd.DataFrame(scaled_fare_age, columns=['fare', 'age'])

# แทนที่คอลัมน์ Fare และ Age ใน DataFrame เดิมด้วยข้อมูลที่ถูกปรับขนาด
titanic[['fare1', 'age1']] = scaled_fare_age_df[['fare', 'age']]
```

```python
titanic
```

```python
#สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)
```

```python
# สร้าง 'AgeGroup' โดยใช้ pd.cut
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=[0, 12, 18, 50, 100],
                             labels=['Child', 'Teenager', 'Adult', 'Elder'], right=False)
```

```python
titanic
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ตั้งค่าขนาดของกราฟ
plt.figure(figsize=(15, 5))

# ฮิสโตแกรมสำหรับ Age แสดงการกระจายของอายุผู้โดยสาร โดยมีเส้น KDE (Kernel Density Estimation) เพื่อแสดงความหนาแน่นของการกระจาย
plt.subplot(1, 3, 1)
sns.histplot(titanic['age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
```

จากกราฟที่แสดงการกระจายของอายุผู้โดยสาร (Age Distribution)  พบว่าการกระจายอายุมีลักษณะเบ้ขวาเล็กน้อย โดยมีผู้โดยสารจำนวนมากอยู่ในช่วงอายุ 20-30 ปี

```python
# แผนภาพกล่องสำหรับ Fare แสดงการกระจายของค่าโดยสาร โดยแสดงค่ามัธยฐาน, ควอไทล์, และค่าผิดปกติ
plt.subplot(1, 3, 2)
sns.boxplot(x=titanic['fare'])
plt.title('Fare Distribution')
plt.xlabel('Fare')
```

จากแผนภาพกล่องสำหรับค่าโดยสาร (Fare Distribution) พบว่าค่าโดยสารส่วนใหญ่มีค่าอยู่ในช่วงต่ำ
โดยมีค่าผิดปกติ (outliers) อยู่ที่ค่าโดยสารสูงๆ ซึ่งอาจเป็นผู้โดยสารที่เดินทางในชั้นโดยสารสูง หรือจ่ายค่าโดยสารพิเศษ

```python
# แผนภาพแท่งสำหรับ Pclass แสดงจำนวนผู้โดยสารในแต่ละชั้นโดยสาร
plt.subplot(1, 3, 3)
sns.countplot(x=titanic['pclass'])
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')

plt.tight_layout()
plt.show()
```

จากแผนภาพแท่งสำหรับชั้นโดยสาร (Passenger Class Distribution) พบว่ามีผู้โดยสารส่วนใหญ่เดินทางในชั้นโดยสารที่ 3 (Pclass = 3) ซึ่งเป็นชั้นโดยสารที่ถูกที่สุด และมีผู้โดยสารในชั้นโดยสารที่ 1 และ 2 น้อยกว่า

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
#i. ความสัมพันธ์ระหว่าง survived และ pclass
#ผู้โดยสารในชั้น 1 (pclass = 1) มีอัตราการรอดชีวิตสูงที่สุด รองลงมาคือชั้น 2 และชั้น 3 ตามลำดับ แผนภูมิแท่งแสดงให้เห็นถึงความสัมพันธ์นี้ได้อย่างชัดเจน
print(titanic.groupby('pclass')['survived'].mean())

# แผนภูมิแท่ง
sns.barplot(x='pclass', y='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#ii. ความสัมพันธ์ระหว่าง survived และ sex
#ผู้หญิงมีอัตราการรอดชีวิตสูงกว่าผู้ชายอย่างมาก แผนภูมิแท่งแสดงให้เห็นถึงความแตกต่างนี้ได้อย่างชัดเจน
print(titanic.groupby('sex')['survived'].mean())

# แผนภูมิแท่ง
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#iii. ความสัมพันธ์ระหว่าง survived และ age
#แผนภาพกระจายแสดงให้เห็นว่าเด็กเล็กมีอัตราการรอดชีวิตสูงกว่าผู้ใหญ่วัยกลางคนเล็กน้อย อย่างไรก็ตาม ความสัมพันธ์ระหว่างอายุและการรอดชีวิตนั้นไม่ชัดเจนเท่ากับความสัมพันธ์ระหว่าง pclass หรือ sex และการรอดชีวิต
# แผนภาพกระจาย
sns.violinplot(x='survived', y='age', data=titanic)
plt.title('Survival Rate by Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
#iv. ความสัมพันธ์ระหว่างจำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare
#แผนภาพกระจายแสดงให้เห็นว่าไม่มีความสัมพันธ์ที่ชัดเจนระหว่างจำนวนสมาชิกครอบครัวที่รอดชีวิตและค่าโดยสาร สัมประสิทธิ์สหสัมพันธ์ที่ใกล้เคียงกับศูนย์ยืนยันว่าไม่มีความสัมพันธ์เชิงเส้นที่แข็งแกร่งระหว่างตัวแปรทั้งสอง
# สร้างคอลัมน์ใหม่สำหรับจำนวนสมาชิกครอบครัวที่รอดชีวิต
titanic['family_survived'] = titanic['sibsp'] + titanic['parch']

# แผนภาพกระจาย
sns.scatterplot(x='family_survived', y='fare', data=titanic)
plt.title('Family Size vs Fare')
plt.xlabel('Number of Family Members Survived')
plt.ylabel('Fare')
plt.show()

# สัมประสิทธิ์สหสัมพันธ์
print(titanic['family_survived'].corr(titanic['fare']))
```

* (10 points) สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร ทำการสรุปผลที่ได้จากตาราง

```python
from IPython.display import display

summary_table = titanic.groupby(['survived', 'pclass', 'sex', 'AgeGroup'])['fare'].agg(['count', 'mean', 'median'])

display(summary_table)
```

จากการตาราง สามารถสรุปได้ว่าปัจจัยต่างๆ เช่น ชั้นโดยสาร, เพศ, และอายุ มีผลต่อโอกาสการรอดชีวิตของผู้โดยสารเรือไททานิค ผู้โดยสารที่เป็นผู้หญิง, อยู่ในชั้นโดยสารสูง, และเป็นเด็กมีโอกาสรอดชีวิตสูงกว่าผู้โดยสารกลุ่มอื่นๆ
