````
import polars as pl

# สร้าง DataFrame ตัวอย่าง
data = {
    "type": ["มุขตลก"] * 50 + ["อื่นๆ"] * 176,
    "value": range(226)
}
folk_df = pl.DataFrame(data)

# กรองเฉพาะแถวที่มีค่า 'type' เท่ากับ 'มุขตลก'
joke_df = folk_df.filter(pl.col('type') == 'มุขตลก')

# สร้าง list of logics
# เลือก 10 ตัวแรก
first_10_indices = joke_df.head(10).select(pl.col('value')).to_series().to_list()

# เลือก 15 ตัวสุดท้าย
last_15_indices = joke_df.tail(15).select(pl.col('value')).to_series().to_list()

# รวมดัชนีของ 10 ตัวแรกและ 15 ตัวสุดท้ายเข้าด้วยกัน
selected_indices = set(first_10_indices + last_15_indices)

# สร้าง Boolean series ที่แสดงว่าแถวใดใน DataFrame หลักถูกเลือก
result_series = pl.Series([i in selected_indices for i in range(len(folk_df))])

print(result_series)
````

````
sum = 0
N = 0
for a in female['age']:

    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง {sum/N}')
````

````
female_KK = female[female['province_of_onset']=='ขอนแก่น']

sum = 0
N = 0
for a in female_KK['age']:

    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง ในจังหวัดขอนแก่น {sum/N}')
````

### การวนลูป record ในตาราง  (```.iterrows```)

````
this_data = data_covid[['sex','age','province_of_onset']]
````

````
for each_row in this_data.iterrows():
    # print(each_row)
    # print(each_row[1])
    if (each_row[1]['age'] == 20) and (each_row[1]['province_of_onset'] == 'ขอนแก่น'):
        print(each_row)
````

### การวนลูป แบบมองตารางแพนด้าส์(pandas dataframe) เป็น numpy array  (```.iloc```)

````
for each_row in range(this_data.shape[0]):
    if (this_data.iloc[each_row,1] == 20) and (this_data.iloc[each_row,2] == 'ขอนแก่น'):  #each_row คือ ตัวเลข
        print(each_row)
        print(this_data.iloc[each_row,:])
````
